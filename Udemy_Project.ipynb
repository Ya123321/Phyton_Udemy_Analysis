```python
# There is a low correlation between the length of the course&the number of lectures and reviews&comments
# There is a preference to purchase a course that is not too long, even though it is possible to learn more from a longer course
corr=df[['content_length_min','num_lectures','num_reviews','num_comments']].corr()
print(corr)
```

```python
sns.heatmap(data=corr,cmap="Blues", annot=True)
```

In order to strengthen my skills, I want to take a Python course in English, to last between 5-30 hours, and it won't cost more than $50

```python
# The topic of the course should be Python
df_py=df.query("topic in ('Python')") 
df_py
```

```python
# The course should be for data analyst
D="Data Analysis"
df_filter=df_py["title"].str.contains(D)
df_py=df_py.loc[df_filter,["id","title","avg_rating","language","price","content_length_min","instructor_name","num_subscribers"]]
df_py
```

```python
# Language, Duration and Price Preferences
df_py=df_py[df_py["language"] == 'English']
df_py=df_py[df_py["price"]<=50]
df_py=df_py.query("content_length_min <= 1800 & content_length_min >= 300")
# reset index for the new data frame
df_py=df_py.reset_index(drop=False)
```

```python
# Check who is the most recommended instructor by avarge rating from the lines I filtered
Top_5_Phyton_instructor_name=df_py.groupby("instructor_name").sum().sort_values(by="avg_rating", ascending=False)["avg_rating"]
Top_5_Phyton_instructor_name
```

```python
# Check which course has the most subscribers from the lines I filtered
my_ideal_Phyton_course=df_py.sort_values("num_subscribers",ascending=False)
my_ideal_Phyton_course
```
